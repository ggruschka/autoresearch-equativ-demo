# CTR Prediction — Experiment Progress

Visualizes autoresearch experiment results from `results.tsv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
df = pd.read_csv("results.tsv", sep="\t", names=["commit", "val_logloss", "val_auc", "status", "description"])
print(f"Total experiments: {len(df)}")
print(f"Kept: {(df.status == 'kept').sum()} | Regressed: {(df.status == 'regressed').sum()} | Crashed: {(df.status == 'crashed').sum()}")
df.tail(10)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot val_logloss over time (lower is better)
kept = df[df.status == "kept"].reset_index(drop=True)
regressed = df[df.status == "regressed"].reset_index(drop=True)

ax1.plot(kept.index, kept.val_logloss, "g.-", label="Kept", markersize=8)
if len(regressed) > 0:
    ax1.scatter(regressed.index, regressed.val_logloss, c="red", marker="x", s=40, label="Regressed", zorder=5)
ax1.set_xlabel("Experiment #")
ax1.set_ylabel("Validation Log-Loss (lower = better)")
ax1.set_title("CTR Prediction: val_logloss")
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot val_auc over time (higher is better)
ax2.plot(kept.index, kept.val_auc, "b.-", label="Kept", markersize=8)
if len(regressed) > 0:
    ax2.scatter(regressed.index, regressed.val_auc, c="red", marker="x", s=40, label="Regressed", zorder=5)
ax2.set_xlabel("Experiment #")
ax2.set_ylabel("Validation AUC (higher = better)")
ax2.set_title("CTR Prediction: val_auc")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("progress.png", dpi=150)
plt.show()
print(f"Best val_logloss: {kept.val_logloss.min():.4f}")
print(f"Best val_auc: {kept.val_auc.max():.4f}")
```

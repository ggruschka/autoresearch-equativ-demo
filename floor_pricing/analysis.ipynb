# Floor Pricing — Experiment Progress

Visualizes autoresearch experiment results from `results.tsv`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
df = pd.read_csv("results.tsv", sep="\t", names=["commit", "total_revenue", "fill_rate", "status", "description"])
print(f"Total experiments: {len(df)}")
print(f"Kept: {(df.status == 'kept').sum()} | Regressed: {(df.status == 'regressed').sum()} | Crashed: {(df.status == 'crashed').sum()}")
df.tail(10)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot total_revenue over time (higher is better)
kept = df[df.status == "kept"].reset_index(drop=True)
regressed = df[df.status == "regressed"].reset_index(drop=True)

ax1.plot(kept.index, kept.total_revenue, "g.-", label="Kept", markersize=8)
if len(regressed) > 0:
    ax1.scatter(regressed.index, regressed.total_revenue, c="red", marker="x", s=40, label="Regressed", zorder=5)
ax1.set_xlabel("Experiment #")
ax1.set_ylabel("Total Revenue ($, higher = better)")
ax1.set_title("Floor Pricing: Revenue")
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot fill_rate over time
ax2.plot(kept.index, kept.fill_rate, "b.-", label="Kept", markersize=8)
ax2.axhline(y=0.40, color="red", linestyle="--", alpha=0.5, label="Min fill rate (0.40)")
if len(regressed) > 0:
    ax2.scatter(regressed.index, regressed.fill_rate, c="red", marker="x", s=40, label="Regressed", zorder=5)
ax2.set_xlabel("Experiment #")
ax2.set_ylabel("Fill Rate")
ax2.set_title("Floor Pricing: Fill Rate")
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("progress.png", dpi=150)
plt.show()
print(f"Best total_revenue: ${kept.total_revenue.max():,.2f}")
print(f"Fill rate at best: {kept.loc[kept.total_revenue.idxmax(), 'fill_rate']:.4f}")
```
